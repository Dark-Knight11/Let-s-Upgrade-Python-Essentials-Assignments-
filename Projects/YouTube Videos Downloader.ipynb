```python
from pytube import YouTube # package for downloading YouTube videos
import os # package for saving videos in a directory
from tkinter import * # package for creating GUI

root = Tk() # main window start
root.geometry('1280x720') # resolution of window
root.configure(bg="black") # background color of window
root.title('Youtube Video Downloader') # Title of GUI window

# Labels inside the window
Label_2 = Label(root, text="YouTube", fg = "white", bg = "red", font = ("bold", 30))
Label_2.place(x = 535, y = 30) # their position in window
Label_1 = Label(root, text = "Paste The Youtube Link Here", bg="black", fg="white", font = ("bold",20))
Label_1.place(x = 430,y = 90) 

mylink = StringVar() # input inside box will be stored in this variable
link = Entry(root, width=60, textvariable=mylink) # input box for pasting the video link
link.place(x = 430, y = 140) # its position

# downloading video code
def vidDwl():
    a = mylink.get() # stores input of link in this variable
    vid = YouTube(a).streams.filter(progressive=True, file_extension='mp4').order_by('resolution').desc().first() # pytube code
    # checks if video is downloadable or not through pytube 
    if len(YouTube(a).streams) == 0: 
        display = (vid.title + " can't be downloaded")
        Label_3 = Label(root, text = display,bg="black", fg="white", font = ("normal",12))
        Label_3.place(x=10, y=220)
        print (vid.title, " can't be downloaded")
    # if downloadable then downloads and stores it in E:/Letsupgrade directory
    else:
        # displays successfull msg
        dis = (vid.title + " is downloaded at E:/Letsupgrade") 
        Label_4 = Label(root, text = dis, bg="black", fg="white", font = ("normal",12))
        Label_4.place(x=10, y=220)
        print (vid.title, " is downloaded at E:/Letsupgrade")
        # checks if folder exists or not
        if not os.path.exists('E:/letsupgrade'):
            os.makedirs('E:/Letsupgrade') # creates folder if it is not there
        vid.download('E:/Letsupgrade') # downloads video in that folder

# Download button which executes vidDwl function
b1 = Button(root,text="Download Video", width=30, bg='green',fg="white", command=vidDwl, font = ("bold", 13))
b1.place(x=480, y=175)
root.mainloop() # main window end
```
